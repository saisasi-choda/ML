```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#fig=plt.figure()
plt.figure()
```

```python
x=np.arange(1,1000)
y=x**4
a=np.arange(1,1000)
b=a
plt.plot(x,y)
plt.show()
```

```python
fig=plt.figure() #create a fig
#BIG PIC
axes1=fig.add_axes([0,0,1,1]) # parameter should be list , define your axes
#plot your dependent variables on the axes
axes1.set_xlim(0,10)
axes1.set_ylim(0,1000)
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('x square-main ')
axes1.plot(x,y)
#small pic
axes2=fig.add_axes([0.6,0.4,0.30,0.30])
axes2.set_xlim(0,10)
axes2.set_ylim(0,10)
axes2.set_xlabel('x')
axes2.set_ylabel('y')
axes2.set_title('a linear b ')
axes2.plot(a,b);
```

```python
fig=plt.figure(figsize=(2,3),dpi =100)
axes1=fig.add_axes([0,0,1,1])
axes1.plot(a,b)
fig.savefig('my_picture.png',dpi=200,bbox_inches='tight')
```

### **subplots**
subplots function unpacks both fig and axes , the default values is 1,1


```python
fig,axes=plt.subplots()
```

```python
fig,axes=plt.subplots(nrows=1,ncols=3)

for ax in axes:
    ax.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=3,ncols=1)
#fig.dpi=200
for ax in axes:
    ax.plot(x,y)

plt.tight_layout()
```

```python
fig,axes=plt.subplots(nrows=2,ncols=2)
for ax in axes.flat:
    ax.plot(x,y)

for ax in axes.flat:
  ax.set_xlabel('x')
  ax.set_ylabel('y')


for i in range(0,len(axes)):
  for j in range(0,len(axes)):
    axes[i][j].set_title(f'title{i},{j}'.format(i,j))

fig.suptitle("Figure")

fig.tight_layout()
fig.savefig("fourbyfour.png",bbox_inches='tight',dpi=200)
```
