### GROUPBY

```python
df=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/udemy ML/03-Pandas/mpg.csv")
df.head()
```

```python
df.groupby("model_year")#unless you use a aggregation fucntions , group by does not apply its operation on data frame which is called lazy evaluation
```

```python
df.groupby("model_year").mean() #model year will be set to index for new df
```

```python
df.groupby("model_year").mean()["mpg"] #slicing for a single column
```

```python
df.groupby(["model_year","cylinders"]).mean() # pass as list for groupby argument for necessary group by column operations
```

```python
yr_cy=df.groupby(["model_year","cylinders"]).mean()
yr_cy.index.levels#to get the multi level index in df on high level but not exact meaning, 81 outer doesnt mean to have all the inner level index for it
```

```python
yr_cy.index
```

```python
yr_cy.loc[[70,71]]#selecting multiple outer indices
```

```python
yr_cy.loc[[(70,4),(71,4)]] #selecting a row i.e inner index using a outer index with help of tuple input, again if you pass this as a list  it will search at outer index
```

Multi level index operations


```python
yr_cy.xs(key=4,level='cylinders') #selecting at second level of index using xs method
```

the above now if we want to filter the above using muti level say need to have 6,8 values instead of key  4, we cannot give key = 6,8 -> hence we need to prefilter the data and then perform group by

The primary diffrence is it is better to sort out the required data initially before grouping as this make hectic to specify columns after certain grouping operations are performed

```python
df[df['cylinders'].isin([6,8])].groupby(['model_year','cylinders']).mean()
```

```python
yr_cy.swaplevel() #swap the indices level , we can use this based on the situations
```

```python
yr_cy.sort_index(level='cylinders',ascending = False) # inner index sorting is kind of breaking outer level grouping and sorting is found when compared to above df
```

###Aggregation
this is a process of consolidting multiple related columns to get a valuble insights on the specifics

let say we need to check the mean of a column , then we records of that column to get this insight , such a way there are many functions like standard deviation, mean, median,count, which takes multiple records as input and gives one record as output

```python
from numpy.lib.function_base import median
df.agg(['std','mean'])
```

agg() is function that allows us to perform specific operations on columns of dataframe,


```python
operations={'weight':['mean','std','var'],'mpg':['mean','std']}
```

```python
df.agg(operations)
```
