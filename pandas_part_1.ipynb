```python
def convert(num):
  return int(str(num)[-4:])
```

```python
convert(12345734)
```

```python
df["last_four"]=df["CC Number"].apply(convert)
df.describe()
```

```python
def costly(money):
  if money>100:
    return "$$$"
  elif money>50 and money<=100:
    return "$$"
  elif money >0 and money<=50:
    return "$"
```

```python
df["expense_rating"]=df['total_bill'].apply(costly)
df["total_bill"].mean()
```

now if i want to double the total bill amount in the sheet the ideal way is to multily by 2

```python
def quality(bill,tip):
  if tip<0.1*bill:
    return "generous"
  else:
    return "others"

quality(10,0.1)
```

```python
df.head()
```

```python
df["quality"]=df[["total_bill","tip"]].apply(lambda df: quality(df["total_bill"],df["tip"]),axis = 1)
df.head()
```

```python
df["quality"]= np.vectorize(quality)(df["total_bill"],df['tip'])
df.head()
```
